```python
import detector
import pandas as pd
```

```python
d = detector.Detector()
```

```python
names = d.return_all_names()
```

```python
n = []
g = []
h = []
for k, v in names.items():
    n.append(k.lower())
    g.append(sorted([i for i in v.keys()]))
    h.append(' '.join(sorted([i for i in v.keys()])))
```

```python
name_cols = {'name': n, 'gender': g, 'gender_joined': h}
df = pd.DataFrame(name_cols)
```

```python
def matcher(row, term):
    if term in row:
        return 1
    else:
        return 0

for i in ['andy', 'female', 'male', 'most_female', 'mostly_male']:
    df[i] = df.apply(lambda row: matcher(row['gender'], i), axis=1)
```
